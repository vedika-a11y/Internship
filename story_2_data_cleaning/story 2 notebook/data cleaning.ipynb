```python
## Example: drop rows if a key column is missing
# Drop rows where CustomerID or Description is missing
df = df.dropna(subset=['CustomerID', 'Description'])

print("Shape after dropping critical missing rows:", df.shape)
```

```python
#Fill Numerical Missing Values
numeric_cols = ['Quantity', 'UnitPrice']

for col in numeric_cols:
    df[col].fillna(df[col].median(), inplace=True)

print("Missing values after numeric filling:")
df[numeric_cols].isnull().sum()
```

```python
#forward fill
df['InvoiceDate'] = df['InvoiceDate'].ffill()

print("Missing values in InvoiceDate:", df['InvoiceDate'].isnull().sum())
```

```python

df.isnull().sum()
```

```python
## Removing Duplicate Rows
df = df.drop_duplicates()
```

```python
# Handling Invalid Values and Outliers
print("Negative Quantity Count:", (df['Quantity'] < 0).sum())
print("Zero Quantity Count:", (df['Quantity'] == 0).sum())
print("Negative UnitPrice Count:", (df['UnitPrice'] < 0).sum())
print("Zero UnitPrice Count:", (df['UnitPrice'] == 0).sum())
```

```python
df[df['Quantity'] <= 0]
```

```python
df[df['UnitPrice'] <= 0].head()
```

```python
# Remove negative or zero quantity
df = df[df['Quantity'] > 0]

# Remove negative or zero unit price
df = df[df['UnitPrice'] > 0]

print("Shape after removing invalid Quantity and UnitPrice:", df.shape)
```

```python
#IQR Outlier Removal advance cleaning
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return data[(data[column] >= lower) & (data[column] <= upper)]
```

```python
df = remove_outliers_iqr(df, 'Quantity')
df = remove_outliers_iqr(df, 'UnitPrice')

print("Shape after IQR outlier removal:", df.shape)
```

```python
#Standardize Categorical Data
 #Convert text to lowercase, Remove extra spaces, Ensure consistent formatting
df['Description'] = df['Description'].str.strip().str.lower()
df['Country'] = df['Country'].str.strip().str.lower()
df['StockCode'] = df['StockCode'].str.strip()
```

```python
## Converting InvoiceDate to Datetime Format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

print(df['InvoiceDate'].dtype)
```

```python
df['InvoiceDate']
```

```python
#Create Data Quality Report
rows_after = df.shape[0]
cols_after = df.shape[1]
null_after = df.isnull().sum().sum()
duplicates_after = df.duplicated().sum()

report = pd.DataFrame({
    "Metric": ["Rows", "Columns", "Missing Values", "Duplicate Rows"],
    "Before Cleaning": [rows_before, cols_before, null_before, duplicates_before],
    "After Cleaning": [rows_after, cols_after, null_after, duplicates_after]
})

report
```

```python
#missing vakue reduction 
reduction = ((null_before - null_after) / null_before) * 100
print("Missing Values Reduced By:", round(reduction, 2), "%")
```
