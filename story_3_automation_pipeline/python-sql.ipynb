```python
pd.read_sql("""
SELECT SUM(sales) AS Total_Revenue
FROM sales_data
""", engine)
```

```python
#Shows total income generated.
#Helps management understand overall business growth.
```

```python
pd.read_sql("""
SELECT COUNT(DISTINCT invoiceno) AS Total_Orders
FROM sales_data
""", engine)
```

```python
#Shows total number of transactions.
#Helps measure business activity level.
```

```python
pd.read_sql("""
SELECT country, SUM(sales) AS Revenue
FROM sales_data
GROUP BY country
ORDER BY Revenue DESC
""", engine)
```

```python
#Identifies strongest performing countries.
#Helps business focus on profitable markets.
```

```python
pd.read_sql("""
SELECT description, SUM(quantity) AS Total_Quantity_Sold
FROM sales_data
GROUP BY description
ORDER BY Total_Quantity_Sold DESC

""", engine)
```

```python
#Shows most profitable products.
#Important for pricing and marketing strategy
```

```python
pd.read_sql("""
SELECT year, month, SUM(sales) AS Monthly_Revenue
FROM sales_data
GROUP BY year, month
ORDER BY year, month
""", engine)
```

```python
#identifies seasonal trends.
#Helps forecast future sales
```

```python
pd.read_sql("""
SELECT high_value_order, COUNT(*) AS Order_Count
FROM sales_data
GROUP BY high_value_order
""", engine)
```

```python
#Shows how many premium orders exist.
#Helps identify high-value customers.
```

```python
pd.read_sql("""
SELECT AVG(total_sales) AS Avg_Order_Value
FROM sales_data
""", engine)
```

```python
#Shows spending behavior of customers.
#Higher value = better profitability.
```

```python
pd.read_sql("""
SELECT customerid, SUM(sales) AS Total_Spent
FROM sales_data
GROUP BY customerid
ORDER BY Total_Spent DESC
LIMIT 10
""", engine)
```

```python
#Identifies loyal / VIP customers.
#Useful for retention strategies.
```

```python
pd.read_sql("""
SELECT description, AVG(revenue_per_unit) AS Avg_Revenue_Per_Unit
FROM sales_data
GROUP BY description
ORDER BY Avg_Revenue_Per_Unit DESC
LIMIT 10
""", engine)
```

```python
#Identifies premium-priced products.
#Useful for margin optimization.
```

```python
pd.read_sql("""
SELECT COUNT(*) AS Returned_Orders
FROM sales_data
WHERE quantity < 0
""", engine)
```

```python
#Shows return rate.
#High returns = product or quality issue.
```

```python
#Create Function
def get_total_revenue():
    query = "SELECT SUM(sales) AS Total_Revenue FROM sales_data"
    result = pd.read_sql(query, engine)
    return result
```

```python
print(get_total_revenue())
```

```python
def get_top_customers():
    query = """
    SELECT customerid, SUM(sales) AS Total_Spent
    FROM sales_data
    GROUP BY customerid
    ORDER BY Total_Spent DESC
    LIMIT 5
    """
    result = pd.read_sql(query, engine)
    return result
```

```python
print(get_top_customers())
```
