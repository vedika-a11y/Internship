```python
plt.figure()
plt.hist(df["sales"], bins=40)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
#Most sales values are concentrated at the lower end, while a small number of transactions have very high sales values.
#Business Insight  
#The business receives many low-value orders and a few high-value orders. This suggests that increasing average order value could improve total revenue.
```

```python
plt.figure()
plt.boxplot(df["quantity"], vert=False)
plt.title("Quantity Distribution")
plt.xlabel("Quantity")
plt.show()
```

```python
 
#Most orders contain small quantities, but some transactions show very large quantities or negative quantities.

#Business Insight 
#Large orders may represent wholesale or bulk buyers, while negative quantities may indicate returns. .
```

```python
plt.figure()
plt.hist(df["unitprice"], bins=40)
plt.title("Unit Price Distribution")
plt.xlabel("Unit Price")
plt.ylabel("Frequency")
plt.show()
```

```python
 #Most products have low unit prices, while a few products are priced much higher.
 #Business Insight: 
#The product mix includes both affordable and premium products, which may support different customer segments.
```

```python
plt.figure()
plt.scatter(df["quantity"], df["sales"], alpha=0.3)
plt.title("Quantity vs Sales")
plt.xlabel("Quantity")
plt.ylabel("Sales")
plt.show()
```

```python
# presence of the negative values suggests that return or refunds are the part of dataset
#small number of transaction involves large quantities which increases revenue
#Business insights
#sales increases with quantity therfore business relies on customers purchasing multiple items per transactions
#negetive values suggest that return or cancellations exits which may imoact profit margin
```

```python
plt.figure()

corr = df[["quantity","unitprice","sales","revenue_per_unit","total_sales"]].corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Matrix")
plt.show()
```

```python
#sales are stronger by the quantity of purcased 
#unit price have very low impact on the total sales 
#customer buys more items rather than expensives ones

#Business insights
#customer buying larger quantities generate higher sales so bundle offers and quantity discounts have opportunities 
#the strong relationship between quantity and sales indicates that revenue growth is primarily driven by the number of the items sold rather than price of the individual items 
```

```python
monthly_quantity = df.groupby("month")["quantity"].mean()

plt.figure(figsize=(8,5))

monthly_quantity.plot(marker="o")

plt.title("Average Quantity Purchased per Month")
plt.xlabel("Month")
plt.ylabel("Average Quantity")

plt.grid(True)

plt.show()
```

```python
sns.countplot(x="high_value_order", data=df)
plt.yscale("log")
plt.title("High Value Orders Count")
plt.xlabel("High Value Order")
plt.ylabel("Number of Orders")
plt.show()
```

```python
#data is imbalanced and normal value purcase is more, high value orders and less of total transactions 
```

```python
sns.kdeplot(df["revenue_per_unit"], fill=True)

plt.title("Revenue per Unit Distribution")

plt.show()
```

```python
#most transactions have low revenue per unit 
#very few numberr of products generate very high revenue per unit 
#Business Insights
#business depends on may small transactions 
#there arwe some premium products which generate the high profit margin
```

```python
monthly_sales = df.groupby("month")["sales"].sum()

plt.figure()

monthly_sales.plot(kind="pie", autopct="%1.1f%%")

plt.title("Revenue Share by Month")

plt.ylabel("")

plt.show()
```

```python
#november generates the highest revenue and feb and april generates lowest revenue
#Business Insights:
#strong demand during the last months 
#so we can make sales in the months where the revenue is lowest and make production more before the last four months to avoid the shortage.
```
